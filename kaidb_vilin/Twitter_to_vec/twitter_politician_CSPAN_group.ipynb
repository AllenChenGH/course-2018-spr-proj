```python
def group_mems(group):
    mems = []
    for member in tweepy.Cursor(api.list_members, 'CSPAN', group).items():
        mems+= [member]
    json_lst = [ i._json for i in mems]
    mas = [i['screen_name'] for i in json_lst]#  if 'M' in i['location'] ]
    return mas

g = []
for i in lists:
    g += group_mems(i)
    
    
```
